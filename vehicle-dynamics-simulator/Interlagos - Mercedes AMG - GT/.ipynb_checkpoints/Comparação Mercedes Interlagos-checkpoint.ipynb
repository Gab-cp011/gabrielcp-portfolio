```python
import matplotlib.pyplot as plt
import pandas as pd
from loop import *
from mpl_toolkits import mplot3d
import ferramentas
```

```python
data_real = pd.read_csv('mercedesinterlagosfilter.csv')
data_real.speed = data_real.speed*3.6
data_sim = pd.read_csv('data.csv')
data_sim.Speed = data_sim.Speed*3.6
data_sim.Ay = data_sim.Ay*-1
coords = pd.read_csv('interlagosxyz.csv')
data_sim.head()
data_real.head()
```

```python
%matplotlib qt
fig = plt.figure()
#ax.set_aspect('equal')
ax = plt.axes(projection='3d')
#ax.set_aspect('equal')
ax.scatter3D(coords.x, coords.y, coords.z,color='red',marker= '.',);
ax.set_aspect('equal')
ax.set(zticklabels=[])
```

```python
%matplotlib inline
plt.plot(data_real.distance,coords.z[1:],'k',label='Interlagos')
plt.grid()
plt.ylabel('Altitude (m)')
plt.xlabel('Distância (m)')
plt.legend()
```

```python
plt.plot(data_sim.Distance,data_sim.Speed,'b',label='Simulado')
plt.plot(data_real.distance,data_real.speed,'r:',label='Referência')
plt.xlabel('Distância (m)')
plt.ylabel('Velocidade (km/h)')
plt.legend()
plt.grid()
```

```python
print('T real = ' + str(data_real.time[len(data_real)-1]))
print('T sim. = ' + str(data_sim.Time[len(data_sim)-2]))
print(data_sim.Distance[len(data_sim)-1]/data_sim.Time[len(data_sim)-2])
print(data_real.distance[len(data_real)-1]/data_real.time[len(data_real)-1])
```

```python
plt.plot(data_sim.Distance,data_sim.Ay,'b',label='Simulado')
plt.plot(data_real.distance,data_real.ay,'r:',label='Referência')
plt.legend()
plt.xlabel('Distância (m)')
plt.ylabel('Aceleração Lateral (m/s²)')
plt.grid()
```

```python
print(data_sim.Speed.mean())
print(data_sim.Speed.std())
print(data_sim.Speed.max())
print(data_sim.Speed.min())
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2,figsize=(10, 5))
fig.suptitle('Aceleração Lateral e Longitudinal')
ax1.plot(data_sim.Distance,data_sim.Ax,'b',label='Simulado')
ax1.set_title('Aceleração Longidutinal')
ax1.plot(data_real.distance,data_real.ax,'r:',label='Referência')
ax1.set_xlabel('Distância (m)')
ax1.set_ylabel('(m/s²)')
ax2.plot(data_sim.Distance,data_sim.Ay,'b',label='Simulado')
ax2.plot(data_real.distance,data_real.ay,'r:',label='Referência')
ax2.set_xlabel('Distância (m)')
ax2.set_ylabel('(m/s²)')
ax2.set_title('Aceleração Lateral')
ax1.legend()
ax2.legend()
ax1.grid()
ax2.grid()

#fig.show()
```

```python
m = 1640
plt.plot(data_sim.Speed,m*data_sim.Ax,'b.',label='Simulado')
plt.plot(data_real.speed,m*data_real.ax,'r+',label='Referência',alpha = 0.7)
plt.xlabel('Velocidade (km/h)')
plt.ylabel('Força Resultante (N)')
plt.legend()
plt.grid()
```

```python
plt.plot(data_sim.Ay,data_sim.Ax,'b.',label='Simulado')
plt.plot(data_real.ay,data_real.ax,'r+',label='Referência',alpha=0.4)
plt.ylabel('Aceleração Longitudinal (m/s²)')
plt.xlabel('Aceleração Lateral (m/s²)')
plt.grid()
plt.legend()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2,figsize=(10, 5))
fig.suptitle('Marcha')
ax1.plot(data_sim.Distance,data_sim.Gears,'b',label='Simulado')
ax1.set_title('Marcha x Distância')
ax1.plot(data_real.distance,data_real.gear-1,'r:',label='Referência')
ax1.set_xlabel('Distância (m)')
ax1.set_ylabel('Marcha')
ax2.plot(data_sim.Speed,data_sim.Gears,'b.',label='Simulado')
ax2.plot(data_real.speed,data_real.gear-1,'r+',label='Referência',alpha=0.1)
ax2.set_xlabel('km/h')
ax2.set_ylabel('Marcha')
ax2.set_title('Marcha x Velocidade')
ax1.legend()
ax2.legend()
ax1.grid()
ax2.grid()

```
